```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('jsrt_metadata.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
max=df['diagnosis'].value_counts().idxmax()
```

```python
df['diagnosis'].replace(np.nan,max,inplace=True)
```

## 1. Find the total number of patients of each disease (diagnosis) and plot in histogram

```python
plt.hist(df['diagnosis'].value_counts())
```

## 2. Find the total number of patients of each disease (diagnosis) with respect to the disease state(malignant or benign) and plot in a pie chart

```python
df1=df[df['state']=='malignant']
df1.head()
```

```python
df1['diagnosis'].value_counts()
```

```python
#plt.title('Number of patitents of each disease(malignant)')
lbl=['lung cancer','lung cancer(adenocarcinoma)','lung cancer(bronchioloalveolar cell carcinoma)','metastasis of colon cancer','lung cancer(squamous cell carcinoma)','lung cancer(small cell carcinoma)','lung canceror metastasis','lung cancer(adenosquamous carcinoma)','lung cancer(large cell carcinoma)','metastatic lung cancer','metastasis of ovarian cancer','metastasis of uterine cancer','susp. lung metastasis']
plt.pie(df1['diagnosis'].value_counts(),labels=lbl,autopct='%1.1f%%',startangle=60,radius=3)
```

```python
df2=df[df['state']=='benign']
df2.head()
```

```python
df2['diagnosis'].value_counts()
```

```python
#plt.title('Number of patitents of each disease(benign)')
lbl=['tuberculoma','granuloma','unknown','inflammatory mass','pulmonary tuberculosis','Inflammatory mass','sclerosing hemangioma','susp. inflammatory mass','pulmonary hematoma','Cryptococcosis','lung abscess','hamartoma','susp. sclerosing hemangioma','organiging pneumonia','susp.tuberculoma','calcified granuloma','pneumonia','organizing pneumonia','susp. hamartoma']
plt.pie(df2['diagnosis'].value_counts(),labels=lbl,autopct='%1.1f%%',startangle=60,radius=3)
```
